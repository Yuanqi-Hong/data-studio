```python
def kg_converter(unit, im, ex):
    if unit=='g' or unit=='cm':
        return (im/1000), (ex/1000)
    elif unit=='oz':
        return (im*0.02835), (ex*0.02835)
    else:
        return im, ex
```

Note: https://apassionatechie.wordpress.com/2017/12/27/create-multiple-pandas-dataframe-columns-from-applying-a-function-with-multiple-returns/

```python
ivory[['im_q','ex_q']] = ivory.apply(lambda row: pd.Series(kg_converter(row['Unit'],row['Importer reported quantity'],row['Exporter reported quantity'])), axis=1)
```

```python
def get_max_quantity(row):
    return max(row['im_q'], row['ex_q'])
```

```python
ivory['max_quantity'] = ivory.apply(get_max_quantity, axis=1)
```

```python
ivory.head(20)
```

```python
ax = ivory.groupby('Year').max_quantity.sum().plot(kind='bar', color='#EC407A', figsize=(10,600))
ax.invert_yaxis()
ax.xaxis.tick_top()
ax.get_yaxis().get_major_formatter().set_scientific(False)
ax.set_yticks([0,10000,50000,100000,200000,1000000,10000000,30000000])
ax.set_yticklabels([0,'10,000','50,000','100,000','200,000','1,000,000','10,000,000','30,000,000'])
ax.tick_params(which='major',left=False, bottom=False, grid_alpha=0.3)
ax.grid(True, axis='y')
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('output.svg')
```
