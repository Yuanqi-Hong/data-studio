## Scraping [WSJ Data](http://graphics.wsj.com/table/NYRIDER_0420?mod=article_inline)

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
```

Here's how to use `WebDriverWait` (instead of just using `time.sleep()`): http://selenium-python.readthedocs.io/waits.html  
Here I'm not using it because the WSJ page was pretty fast when I scraped it.

```python
driver = webdriver.Chrome()
driver.get('http://graphics.wsj.com/table/NYRIDER_0420?mod=article_inline')
```

```python
# Make it show 100 rows per page.
show_dropdown = driver.find_element_by_xpath('//*[@id="tablecontent"]/div[1]/div[1]/div[2]/div[4]/div[1]/span[2]/span/button')
driver.execute_script('arguments[0].scrollIntoView(true)', show_dropdown)
show_dropdown.click()
show_100 = driver.find_element_by_xpath('//*[@id="tablecontent"]/div[1]/div[1]/div[2]/div[4]/div[1]/span[2]/span/ul/li[4]/a')
show_100.click()
```

```python
data = []
for page in range(5):
    rows = driver.find_elements_by_tag_name('tr')
    for row in rows[1:-1]:
        row_data = {}
        cells = row.find_elements_by_tag_name('td')
        row_data['station_trains'] = cells[0].text
        row_data['borough'] = cells[1].text
        row_data['change_from_2013'] = int(''.join(filter(str.isdigit, cells[2].text)))
        row_data['prc_change_from_2013'] = float(''.join(s for s in cells[3].text if s != '%'))
        row_data['2014_avg_ridership'] = int(''.join(filter(str.isdigit, cells[4].text)))
        row_data['2013_avg_ridership'] = int(''.join(filter(str.isdigit, cells[5].text)))
        row_data['2011_avg_ridership'] = int(''.join(filter(str.isdigit, cells[6].text)))
        row_data['2010_avg_ridership'] = int(''.join(filter(str.isdigit, cells[7].text)))
        row_data['closed_2014'] = cells[8].text
        data.append(row_data)
    next_button = driver.find_element_by_class_name('page-next').find_element_by_tag_name('a')
    driver.execute_script('arguments[0].scrollIntoView(true)', next_button)
    next_button.click()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('WSJ_subway_data_2014.csv', index=False)
```

```python
driver.quit()
```
